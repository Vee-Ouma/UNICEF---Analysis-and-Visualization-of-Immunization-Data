```python
df = ch_wm_df.merge(hh_df, on = ['HH1','HH2'])
```

```python
df.head()
```

```python
df = df.loc[:, ~df.columns.str.startswith('UF')]
df = df.loc[:, ~df.columns.str.startswith('HH')]
df = df.loc[:, ~df.columns.str.startswith('LN')]
df.drop('AN1', axis=1, inplace=True)
```

```python
df['IM1'].unique()
```

```python
#Set the target to column IM1 - Has the card and the card is seen
#Seperate the target column from the df.

df['has_card'] = np.where(df['IM1'] == 1, 1,0)
```

```python
df = df.loc[:, ~df.columns.str.startswith('IM')]
```

```python
df_majority = df[df.has_card == 0]
df_minority = df[df.has_card == 1]
```

```python
print(len(df_majority))
print(len(df_minority))
```

```python
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace= True,     # sample with replacement
                                 n_samples=32289,    # to match majority class
                                 random_state=123) # reproducible results
 
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
 
# Display new class counts
df_upsampled.has_card.value_counts()
```

```python
df = df_upsampled
```

```python
#One hot encoding of categorical values using get_dummies function

dummies = pd.get_dummies(df)
dummies.head()
```

```python
#Replace NaN with 0

dummies = dummies.fillna(0)
```

```python
target = df['has_card']
target.unique()
```

```python
#Creating a binary variable since by coding Yes to 1 and any other response as 0.

y=['has_card']
dummy_vars =dummies.columns.values.tolist()
X=[i for i in dummy_vars if i not in y]
```

```python
#Using a combination of Recursive Feature Selection and Logistic regression
#to select the top 25 features that can predict has_been_vaccinated variable

logreg = LogisticRegression()
rfe = RFE(logreg,n_features_to_select= 25)
rfe = rfe.fit(dummies[X],dummies[y].values.ravel())
print(rfe.support_,rfe.ranking_)
```

```python
#Display column name and numbers of the top 25 features from the model.

cols = []
for i in range(len(rfe.ranking_)):
    if rfe.ranking_[i] == 1:
        print(i,dummies.columns[i])   
        cols.append(dummies.columns[i])
```

```python
print ('coefficients',rfe.estimator_.coef_)
```

```python
#Training the logistic regression model

X_train, X_test, y_train, y_test = train_test_split(dummies[cols],dummies[y].values.ravel(), test_size=0.3, random_state=0)
logreg.fit(X_train, y_train)
```

```python
feature_importance = abs(logreg.coef_[0])
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

featfig = plt.figure(figsize=(20,20))
featax = featfig.add_subplot(1, 1, 1)
featax.barh(pos, feature_importance[sorted_idx], align='center')
featax.set_yticks(pos)
featax.set_yticklabels(np.array(dummies[cols].columns)[sorted_idx], fontsize=8)
featax.set_xlabel('Relative Feature Importance')
plt.tick_params(axis='both', which='major', labelsize=15)
plt.tick_params(axis='both', which='minor', labelsize=15)
plt.suptitle('Logistic Regression Feature Importance For Predicting Pass/Fail', fontsize=20)
plt.tight_layout()   
plt.show()
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
#Overall classification report

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#Training the Random Forests model

X_train, X_test, y_train, y_test = train_test_split(dummies[X],dummies[y].values.ravel(), test_size=0.3, random_state=0)
```

```python
# Create a random forest Classifier. By convention, clf means 'Classifier'
rfc = RandomForestClassifier()

# Train the Classifier to take the training features and learn how they relate
# to the training y 
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
print('Accuracy of random forest classifier on test set: {:.2f}'.format(rfc.score(X_test, y_test)))
```

```python
#Overall classification report

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
kfold = model_selection.KFold(n_splits=10, random_state=7)
scoring = 'accuracy'
results = model_selection.cross_val_score(rfc, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```
